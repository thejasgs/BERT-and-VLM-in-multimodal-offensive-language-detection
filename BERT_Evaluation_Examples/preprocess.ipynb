```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import os
```

```python
df_train = pd.read_csv('train.csv')
df_val = pd.read_csv('val.csv')
df_test = pd.read_csv('test.csv')
```

```python
print(df_test['somewhat harmful'].value_counts())
print(df_test['very harmful'].value_counts())
print(df_test['not harmful'].value_counts())
```

```python
df_train_bin = df_train.drop(columns=['id', 'text', 'community', 'individual', 'society', 'organization'], axis=1)
df_val_bin = df_val.drop(columns=['id', 'text', 'community', 'individual', 'society'], axis=1)
df_test_bin = df_test.drop(columns=['id', 'text', 'community', 'individual', 'society'], axis=1)

# Combine harmful indicators into single binary label
# if not harmful == 1 -> label 0, else label 1
for df in (df_train_bin, df_val_bin, df_test_bin):
    df['label'] = np.where(df['not harmful'] == 1, 0, 1)
    df.drop(columns=['very harmful', 'not harmful', 'somewhat harmful'], inplace=True)
```

```python
print(df_test_bin['label'].value_counts())
```

```python
df_train_val = pd.concat([df_train_bin, df_val_bin], ignore_index=True)
print(df_train_val['label'].value_counts())
```

```python
df_all = pd.concat([df_train_val, df_test_bin], ignore_index=True)
print(df_all['label'].value_counts())
```

```python
IMG_DIR = './images/'

# Verify that each image referenced in the dataframe actually exists on disk
all_images = set(df_all['image'].astype(str))
files_on_disk = set(os.listdir(IMG_DIR))
missing_files = sorted(list(all_images - files_on_disk))
if missing_files:
    print("Missing image files:")
    for f in missing_files:
        print(f)
else:
    print("All images are present in the directory.")
```

```python
df_train_val.to_csv('train_val_bin.csv', index=False)
df_test_bin.to_csv('test_bin.csv', index=False)
```
